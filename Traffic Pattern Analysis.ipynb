```python
import pandas as pd

df = pd.read_csv("traffic_pattern_data.csv")
df['Timestamp'] = pd.to_datetime(df['Timestamp'])
df.head()
```

```python
import matplotlib.pyplot as plt

df.set_index('Timestamp', inplace=True)
df['Vehicle_Count'].plot(figsize=(12, 5), title='Traffic Density Over Time')
plt.ylabel("Vehicle Count per Hour")
plt.grid(True)
plt.show()
```

```python
df['Hour'] = df.index.hour
hourly_avg = df.groupby('Hour')['Vehicle_Count'].mean()

plt.figure(figsize=(10, 5))
hourly_avg.plot(kind='line')
plt.title("Average Vehicle Count by Hour")
plt.xlabel("Hour")
plt.ylabel("Average Vehicles")
plt.grid(True)
plt.show()
```

```python
threshold = df['Vehicle_Count'].mean() + 2 * df['Vehicle_Count'].std()
df['Anomaly'] = df['Vehicle_Count'] > threshold

plt.figure(figsize=(12,5))
plt.plot(df.index, df['Vehicle_Count'], label='Traffic')
plt.scatter(df[df['Anomaly']].index, df[df['Anomaly']]['Vehicle_Count'], color='red', label='Anomaly')
plt.legend()
plt.title("Traffic with Anomalies Highlighted")
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
import numpy as np

df.reset_index(inplace=True)
X = df[['Hour']]
y = df['Vehicle_Count']

model = LinearRegression()
model.fit(X, y)

df['Predicted'] = model.predict(X)

plt.figure(figsize=(10,5))
plt.scatter(df['Hour'], df['Vehicle_Count'], alpha=0.3, label='Actual')
plt.plot(df['Hour'], df['Predicted'], color='red', label='Predicted')
plt.xlabel("Hour")
plt.ylabel("Vehicle Count")
plt.title("Regression Model for Traffic Prediction")
plt.legend()
plt.show()
```
